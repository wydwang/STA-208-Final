```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
svm_acc = np.array([42082,828,6173,2620]).reshape(2,2)
svm_acc = pd.DataFrame(svm_acc, 
                     columns = ["Predicted 0", "Predicted 1"],
                     index = ["Observed 0", "Observed 1"])
svm_acc_g = np.array([42097, 813,5911,2882]).reshape(2,2)
svm_acc_g = pd.DataFrame(svm_acc_g, 
                     columns = ["Predicted 0", "Predicted 1"],
                     index = ["Observed 0", "Observed 1"])
```

```python
sns.heatmap(svm_acc, annot = True, fmt = "d", cmap = "YlGn")
```

```python
sns.heatmap(svm_acc_g, annot = True, fmt = "d", cmap = "YlGn")
```
